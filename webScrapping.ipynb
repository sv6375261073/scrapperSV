```python
#  fake message title retrival 
def title_retrive(title):
    title=str(title)
    title_=re.findall('>.*',title)
    title=[str(title).replace('</h1>','').replace('>','') for title in title_]
    return title[0]
# date retrieval 
def date_retrive(date):
    date=str(date)
    date_=re.findall('>.*',date)   #"^ftp:// & .jpg$"
    date=[str(d).replace('</span>','').replace('>','') for d in date_]
    return date[0]
#  multple fake text concatenation  
def con(lst):
    strtxt=""
    for txt in lst:
        strtxt+=txt
    return strtxt
# retrieve fake text which is under quotes only 
def txt_reterive(table):
    strtxt=""
    for txt in table:                                        #    
        content=re.findall('.*',str(txt))
        msg=[content_txt.replace("</p>","").strip().split("") for content_txt in content if content_txt.find("https:") ==-1 ]
        for i in msg:
            if len(i)>0:
                strtxt+=con(i)+"\n"
    return strtxt
```

```python
df=pd.DataFrame(columns=['title','link','date','text','category'])
counter=0
# accessing all links from file and access fake text,date title etc regarding that link 
with open('urls.txt','r') as f:
    for url in f:
        url=str(url).replace('\n','')
        r = requests.get(url) 
        soup = BeautifulSoup(r.content, 'html.parser') 
        table = soup.find_all('p')
        date=soup.find("span",class_="updated")
        title=soup.find('h1',class_="entry-title h1")
        text=txt_reterive(table)
        title=title_retrive(title)
        date=date_retrive(date)
        # making dataframe 
        df.loc[counter]=[title,url,date,text,'fake']
        
```
